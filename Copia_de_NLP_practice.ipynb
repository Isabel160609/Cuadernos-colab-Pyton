```python
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
#from nltk.stem import WordNetLemmatizer (es mas para ingles)
# Si usas stemming, también necesitarías:
#from nltk.stem import PorterStemmer
from nltk.stem.snowball import SnowballStemmer
import nltk

nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('wordnet')

frase_original = "¡Los científicos han descubierto que el 🌞 (sol) emite partículas energéticas a una velocidad de 1.5 millones de kilómetros por hora!"

print(f"Frase original: {frase_original}\n")

# 1. Normalización a minúsculas
frase_minusculas = frase_original.lower()
print(f"1. Después de normalizar a minúsculas: {frase_minusculas}")

# 2. Eliminación de puntuación y emojis
# Usamos expresiones regulares para quitar todo lo que no sea una letra o un espacio
frase_sin_puntuacion = re.sub(r'[^a-zA-Z\s]', '', frase_minusculas)
#frase_sin_puntuacion = [word for word in frase_minusculas if word not in string.punctuation]
print(f"2. Después de eliminar puntuación y emojis: {frase_sin_puntuacion}")

# 3. Eliminación de números
# Aunque ya los quitamos con la regex anterior, si hubieran letras junto a números,
# o si quisiéramos quitar solo números y no otros símbolos, haríamos esto:
frase_sin_numeros = re.sub(r'\d+', '', frase_sin_puntuacion)
print(f"3. Después de eliminar números (en este caso ya se fueron): {frase_sin_numeros}")

# 4. Tokenización
# Primero, tokenizamos la frase ya limpia
tokens = word_tokenize(frase_sin_numeros, language='spanish') # Especificamos el idioma
print(f"4. Después de tokenizar: {tokens}")

# 5. Eliminación de Stopwords
# Obtenemos la lista de stopwords en español
stop_words_es = set(stopwords.words('spanish'))
tokens_sin_stopwords = [palabra for palabra in tokens if palabra not in stop_words_es]
print(f"5. Después de eliminar stopwords: {tokens_sin_stopwords}")

# Opcional: Lematización (mucho más común que Stemming en tareas modernas)
# Para lematizar correctamente, a menudo se necesita conocer la parte del habla (sustantivo, verbo, etc.)
# Pero para un ejemplo básico, podemos lematizar sustantivos por defecto.
lemmatizer = WordNetLemmatizer()
# Como WordNetLemmatizer funciona mejor con tags de POS de WordNet, y en español es más complejo,
# para este ejemplo simple, lematizaremos como sustantivos (n).
# En un proyecto real, usarías un tagger de POS.
tokens_lematizados = [lemmatizer.lemmatize(word) for word in tokens_sin_stopwords]
# Nota: WordNetLemmatizer está entrenado principalmente en inglés.
# Para español, librerías como SpaCy o Stanza son más adecuadas para la lematización.
# Sin embargo, para fines demostrativos de la técnica, sirve.
print(f"6. Después de lematizar (con WordNetLemmatizer, que es más fuerte en inglés): {tokens_lematizados}")

# Opcional: Stemming (menos preciso que lematización, pero más simple)
# stemmer = PorterStemmer() # PorterStemmer es para inglés. Para español, se usa SnowballStemmer.
stemmer_es = SnowballStemmer("spanish")
tokens_stemmed = [stemmer_es.stem(word) for word in tokens_sin_stopwords]
print(f"7. Después de stemming (con SnowballStemmer para español): {tokens_stemmed}")
```

La versión mejorada y más corta

```python
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem.snowball import SnowballStemmer
import nltk

# Aseguramos las descargas necesarias de NLTK
nltk.download('punkt_tab', quiet=True)
nltk.download('stopwords', quiet=True)

frase_original = "¡Los científicos han descubierto que el 🌞 (sol) emite partículas energéticas a una velocidad de 1.5 millones de kilómetros por hora!"

print(f"Frase original: {frase_original}\n" + "-"*50)

# 1. Normalización a minúsculas
frase_minusculas = frase_original.lower()

# 2. Limpieza Inteligente (Puntuación, Emojis y Números en un solo paso)
# Agregamos 'áéíóúüñ' a la regla para que NO elimine los caracteres del español.
# Al NO incluir los números (\d) ni los signos, se van automáticamente.
frase_limpia = re.sub(r'[^a-záéíóúüñ\s]', '', frase_minusculas)
print(f"1. Texto limpio (conservando acentos y Ñ):\n   -> {frase_limpia}\n")

# 3. Tokenización
# Convertimos la cadena de texto limpia en una lista de palabras
tokens = word_tokenize(frase_limpia, language='spanish')
print(f"2. Tokens generados:\n   -> {tokens}\n")

# 4. Eliminación de Stopwords
# Usamos un 'set' (conjunto) porque las búsquedas en conjuntos son ultra rápidas en Python
stop_words_es = set(stopwords.words('spanish'))
tokens_sin_stopwords = [palabra for palabra in tokens if palabra not in stop_words_es]
print(f"3. Sin Stopwords (Palabras con significado real):\n   -> {tokens_sin_stopwords}\n")

# 5. Stemming (Reducción a la raíz en Español)
# Usamos SnowballStemmer configurado explícitamente en "spanish"
stemmer_es = SnowballStemmer("spanish")
tokens_stemmed = [stemmer_es.stem(word) for word in tokens_sin_stopwords]
print(f"4. Resultado final tras el Stemming:\n   -> {tokens_stemmed}")
```

#Ejercicio Extracción de caracteristicas

Pasando de la limpieza de texto a la vectorización, que es el proceso de convertir palabras en números para que la IA pueda "entenderlas".

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
import pandas as pd

# Usaremos una lista de "documentos". En este caso, nuestra frase limpia es un solo documento.
# Pero para ver mejor el efecto, añadiré una segunda frase.
documentos_limpios = [
    "científicos descubierto sol emite partículas energéticas velocidad millones kilómetros hora",
    "partículas sol emite energía mucha" # Un segundo documento para comparar TF-IDF
]

print("Documentos limpios para análisis:\n", documentos_limpios)
print("-" * 50)

### 1. Bag-of-Words (BoW) con CountVectorizer

print("--- Bag-of-Words (BoW) ---")

# Creamos una instancia de CountVectorizer
# Esta herramienta aprende el vocabulario de tus documentos y cuenta las ocurrencias.
vectorizador_bow = CountVectorizer()

# Ajustamos el vectorizador a nuestros documentos (aprende el vocabulario)
# y transformamos los documentos en una matriz de conteo de palabras.
matriz_bow = vectorizador_bow.fit_transform(documentos_limpios)

# Obtenemos las palabras que forman el vocabulario (las columnas de nuestra matriz)
nombres_caracteristicas_bow = vectorizador_bow.get_feature_names_out()

# Convertimos la matriz en un DataFrame de Pandas para que sea más fácil de leer
df_bow = pd.DataFrame(matriz_bow.toarray(), columns=nombres_caracteristicas_bow)
df_bow.index = [f"Doc {i+1}" for i in range(len(documentos_limpios))]

print("\nMatriz BoW (Frecuencia de palabras):")
print(df_bow)
print("\nInterpretación: Cada fila es un documento, cada columna es una palabra del vocabulario, y los valores son las veces que la palabra aparece en el documento.")
print("-" * 50)

### 2. TF-IDF con TfidfVectorizer

print("--- TF-IDF ---")

# Creamos una instancia de TfidfVectorizer
# Esta herramienta hace lo mismo que CountVectorizer, pero luego aplica el cálculo TF-IDF.
vectorizador_tfidf = TfidfVectorizer()

# Ajustamos y transformamos los documentos
matriz_tfidf = vectorizador_tfidf.fit_transform(documentos_limpios)

# Obtenemos las palabras que forman el vocabulario
nombres_caracteristicas_tfidf = vectorizador_tfidf.get_feature_names_out()

# Convertimos la matriz en un DataFrame de Pandas
df_tfidf = pd.DataFrame(matriz_tfidf.toarray(), columns=nombres_caracteristicas_tfidf)
df_tfidf.index = [f"Doc {i+1}" for i in range(len(documentos_limpios))]

print("\nMatriz TF-IDF (Importancia de palabras):")
print(df_tfidf)
print("\nInterpretación: Los valores altos indican palabras importantes y distintivas para ese documento, mientras que los valores bajos indican palabras comunes o menos relevantes.")
print("-" * 50)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline # Para encadenar pasos


# Paso 1: Preparar los Datos
# Creamos un conjunto de documentos de ejemplo y sus etiquetas.
# Simplicaremos los documentos para el ejemplo y les asignaremos una "categoría".
# En un escenario real, tendrías muchos más documentos y categorías.

# Documentos "limpios" (ya preprocesados, como en los ejemplos anteriores)
documentos = [
    "científicos descubierto sol emite partículas energéticas velocidad millones kilómetros hora", # Podríamos llamarlo "Ciencia Espacial"
    "partículas sol emite energía mucha",                                                    # Podríamos llamarlo "Conceptos Generales"
    "velocidad luz partículas descubiertas",                                                  # Otro documento de "Ciencia Espacial"
    "energía solar es mucha",                                                                # Otro documento de "Conceptos Generales"
]

# Etiquetas correspondientes a cada documento
# 0 para "Ciencia Espacial", 1 para "Conceptos Generales"
etiquetas = [0, 1, 0, 1]

print("Documentos de entrenamiento:")
for i, doc in enumerate(documentos):
    print(f"  Doc {i+1} (Categoría: {etiquetas[i]}): {doc}")
print("-" * 50)

# Paso 2 & 3: Encadenar la Extracción de Características y el Entrenamiento del Modelo
# Usaremos un "Pipeline" de Scikit-learn, que es muy útil para encadenar pasos.
# Primero, el TfidfVectorizer para convertir el texto a números.
# Segundo, el MultinomialNB (Clasificador Naive Bayes) para el aprendizaje.

modelo_clasificador = Pipeline([
    ('tfidf', TfidfVectorizer()), # Este paso transforma el texto
    ('clf', MultinomialNB()),     # Este paso es el clasificador
])

# Entrenamos el modelo con nuestros documentos y sus etiquetas.
# El pipeline se encarga de que TfidfVectorizer haga su trabajo y luego el clasificador aprenda.
print("Entrenando el modelo...")
modelo_clasificador.fit(documentos, etiquetas)
print("¡Modelo entrenado exitosamente!")
print("-" * 50)

# Paso 4: Hacer una Predicción
# Vamos a probar el modelo con un nuevo texto que no ha visto antes.
nuevo_texto_1 = "sol emite velocidad luz" # Esperamos que sea 0 (Ciencia Espacial)
nuevo_texto_2 = "mucha energía"         # Esperamos que sea 1 (Conceptos Generales)
nuevo_texto_3 = "descubrimiento de partículas" # Esperamos que sea 0 (Ciencia Espacial)

textos_nuevos = [nuevo_texto_1, nuevo_texto_2, nuevo_texto_3]

print("Realizando predicciones sobre nuevos textos:")
predicciones = modelo_clasificador.predict(textos_nuevos)

# Mapeamos las etiquetas numéricas a nombres más legibles para nosotros
mapa_categorias = {0: "Ciencia Espacial", 1: "Conceptos Generales"}

for i, texto in enumerate(textos_nuevos):
    categoria_predicha = mapa_categorias[predicciones[i]]
    print(f"  Texto: '{texto}'")
    print(f"  Predicción: {categoria_predicha}")
    print("-" * 20)

print("\n¡Hemos construido y probado un pequeño modelo de clasificación!")
print("Recuerda que con tan pocos datos, el modelo es muy simple, pero el proceso es el mismo para modelos más complejos.")
```
