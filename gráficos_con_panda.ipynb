```python
import pandas as pd
import matplotlib.pyplot as plt

# Crear un DataFrame con los datos
datos = {
    'Día': ['Lun', 'Mar', 'Mie', 'Jue', 'Vie', 'Sab', 'Dom'],
    'Temp': [22, 24, 23, 28, 26, 30, 31]
}
# Graficar la temperatura promedio por día
df = pd.DataFrame(datos)
print(df)
#plt.plot(df["Día"], df["Temp"], marker= "D", color="red")
plt.plot(df['Día'], df['Temp'], marker='o',color='orange')
plt.title('Evolución Temperatura')
plt.xlabel('Días de la semana')
plt.ylabel('Temperatura en grados celsius')
plt.show()

```
