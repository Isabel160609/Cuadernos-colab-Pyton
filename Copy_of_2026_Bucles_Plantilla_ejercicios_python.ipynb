#**Ejercicios con Bucles (For / While)**

**1. El contador simple**

Enunciado: Usa un bucle for y la función range() para imprimir los números del 1 al 5.

```python
for x in range(1,5+1):
  print(x)
```

**2. Multiplicador de lista**

Enunciado: Dada la lista de números [1, 2, 3, 4], crea un bucle que multiplique cada número por 10 e imprima el resultado.

```python
números = [1, 7, 5, 4]
for x in números:
  print(x*10)
```

**3. Buscador de nombres largos**

Enunciado: Recorre la lista ["Ana", "Sebastian", "Luz", "Alejandra"]. Si el nombre tiene más de 5 letras, imprímelo. (Tip: usa len(nombre)).

```python
nombres = ["Ana", "Sebastian", "Luz", "Alejandra"]
for nombre in nombres:
  if len(nombre) > 5: #verifica la condición de longitud
    print(nombre)
```

**4. La cuenta atrás (While)**

Enunciado: Crea una variable energia = 5. Mientras la energía sea mayor a 0, imprime "Corriendo..." y resta 1 a la energía. Al final imprime "¡Me cansé!".

```python
energia = 5
while energia > 0:
  print("Corriendo...")
  energia -= 1
print("¡Me cansé!")
```

**5. Filtrar pares**

Enunciado: Recorre los números del 1 al 10. Si el número es par, imprímelo. (Tip: un número es par si numero % 2 == 0).

```python
for x in range(1,10+1):
  if x % 2 == 0:
    print(x)
```

**Enunciado:**

Crea un diccionario con 3 colores en inglés y su traducción al español.

*   Usa un bucle for para imprimir solo las claves (los nombres en inglés).
*   Usa un bucle for para imprimir frases como: "The color [key] in Spanish is [value]".





```python
colores = {"red":"rojo", "blue":"azul", "green":"verde"}
for key in colores:
  print(key)
for key, value in colores.items():
    print(f"The color {key} in Spanish is {value}")
```

**Enunciado:** Dada la palabra **"programacion"**, crea un diccionario que cuente cuántas veces aparece cada letra.

**Pista:** Usa un bucle para recorrer la palabra y el método .get() para inicializar el contador de cada letra.

```python
palabra = "programacion"
conteo_letras = {}

for letra in palabra:
    conteo_letras[letra] = conteo_letras.get(letra, 0) + 1

print(conteo_letras)
```

**El Carrito de la Compra**

**Enunciado:** Crea una lista vacía llamada carrito.

1. Pide al usuario que escriba 3 productos y añádelos uno a uno a la lista.

2. Muestra la lista ordenada alfabéticamente.

3.  Elimina el último producto que se añadió.

```python
carrito = []
for x in range(3):
  producto = input("Introduce un producto:")
  carrito.append(producto)
carrito.sort()
print(f"La lista completa: {carrito}")
carrito.pop()
print(f"La lista sin el ultimo producto: {carrito}")
```
