 La Pandilla de las Manzanas 🍏🍎
Imagina que un grupo de amigos quiere saber más sobre su cosecha de manzanas. Cosechan 10 manzanas y quieren analizar su peso para ver cómo es su producción este año.

📊 Queremos calcular:

Media (peso promedio de las manzanas)
Mediana (peso central)
Moda (peso más común)
Desviación estándar (qué tanto varían los pesos)

```python
import numpy as np
import statistics as stats

# Pesos de las manzanas en gramos 🍏🍎
manzanas = [150, 180, 200, 170, 150, 190, 200, 210, 150, 180]

# Estadísticas descriptivas 📊
media = np.mean(manzanas)  # Promedio
mediana = np.median(manzanas)  # Valor central
moda = stats.mode(manzanas)  # Valor más común
desviacion = np.std(manzanas)  # Variabilidad

# Resultados 🍏📊
print(f"📊 Análisis de la cosecha de manzanas 🍎")
print(f"Media: {media}g")
print(f"Mediana: {mediana}g")
print(f"Moda: {moda}g")
print(f"Desviación estándar: {desviacion:.2f}g")
```

```python
import numpy as np
import statistics as stats
import matplotlib.pyplot as plt

# Pesos de las zanahorias en gramos
zanahorias = [300, 280, 290, 310, 275, 290, 295, 315, 290, 280, 310, 305]

# Estadísticas descriptivas 📊
media = np.mean(zanahorias)  # Promedio
mediana = np.median(zanahorias)  # Valor central
moda = stats.mode(zanahorias)  # Valor más común
desviacion = np.std(zanahorias)  # Variabilidad

# Resultados 📊
print(f"📊 Análisis de la cosecha de zanahorias ")
print(f"Media: {media}g")
print(f"Mediana: {mediana}g")
print(f"Moda: {moda}g")
print(f"Desviación estándar: {desviacion:.2f}g")

plt.hist(zanahorias, bins=8, edgecolor='black')
plt.xlabel('Peso (g)')
plt.ylabel('Frecuencia')
plt.title('Distribución de los pesos de las zanahorias')
plt.show()

```
