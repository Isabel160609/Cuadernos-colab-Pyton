5. **Relación Estatura-Peso:** Usa un gráfico de dispersión (scatter) con 5 puntos de datos inventados para ver si hay una relación visual entre la altura de una persona y su peso.

```python
import matplotlib.pyplot as plt
altura = [165,180,160,170,190]
peso = [60,85,50,120,100]
imm = []
for i in range(len(altura)):
    # Convertimos la altura a metros
    altura_metros = altura[i] / 100

    # Aplicamos la fórmula: peso / altura^2
    imc = peso[i] / (altura_metros ** 2)
    imm.append(imc)
print(imm)
# Crear el gráfico de dispersión
tamaño_puntos = [i * 5 for i in imm]
plt.scatter(altura, peso, s=tamaño_puntos, edgecolors='black', marker="s")
plt.title('Relación Estatura-Peso')
plt.xlabel('Altura')
plt.ylabel('Peso')
plt.grid(True)
plt.show()
```

6. Imagina que quieres analizar cómo gastas tu tiempo en un día de 24 horas. Este es un caso de uso clásico para un gráfico de sectores.

```python
import matplotlib.pyplot as plt

# 1. Tus datos originales
actividades = ['Dormir', 'Trabajo/Estudio', 'Transporte', 'Ocio/Gimnasio', 'Tareas del Hogar']
horas = [8, 8, 2, 4, 2]

# 2. Definimos una lista con un color diferente para cada una de las 5 actividades
# Puedes usar nombres de colores en inglés o códigos HEX (como #FF5733)
colores_personalizados = ['skyblue', 'salmon', 'lightgreen', 'gold', 'orchid']

# 3. Creamos el gráfico de quesito
plt.figure(figsize=(6, 6)) # Hacemos la ventana cuadrada para que el queso sea un círculo perfecto

plt.pie(
    horas,                      # Los valores numéricos (el tamaño de cada porción)
    labels=actividades,         # Los nombres de cada porción
    colors=colores_personalizados, # ¡Aquí asignamos los colores diferentes!
    autopct='%1.1f%%',          # Esto calcula y pone el porcentaje automáticamente
    startangle=140,             # Rota el gráfico para una mejor perspectiva visual
    shadow = True,
    wedgeprops={'edgecolor': 'white', 'linewidth': 2} # Añade una línea blanca elegante entre los quesitos
)

# 4. Título del gráfico
plt.title("Distribución de mi Tiempo Diario (24 Horas)", fontsize=14, fontweight='bold')

# 5. Mostramos el resultado
plt.show()
```

7. **Distribución de Exámenes:** Imagina que tienes una lista de 20 notas de alumnos (del 1 al 10). Crea un histograma para ver qué notas son las más frecuentes. notas = [1, 5, 8, 9, 7, 10, 4, 6, 8, 8, 9, 5, 3, 7, 8, 10, 2, 9, 8, 7]

```python
import matplotlib.pyplot as plt

notas = [1, 5, 8, 9, 7, 10, 4, 6, 8, 8, 9, 5, 3, 7, 8, 10, 2, 9, 8, 7]

plt.figure(figsize=(10, 5))

# 1. Definimos los cortes de las barras exactos desde el 1 hasta el 11
# (ponemos 11 para que el contenedor de la nota 10 se dibuje bien)
rango_notas = range(1, 12)

plt.hist(
    notas,
    bins=rango_notas,         # <-- CAMBIO 1: Forzamos barras individuales para cada nota
    color="mediumpurple",
    edgecolor="indigo",
    linewidth=1,
    alpha=0.85,
    rwidth=0.85,              # <-- RECOMENDACIÓN: Separamos las barras para ver bien cada nota
    cumulative=False
)

plt.title("Distribución de Notas del Examen", fontsize=14, fontweight='bold')
plt.xlabel("Notas")
plt.ylabel("Número de Alumnos")

# 2. CAMBIO 2: Centramos los números debajo de cada barra correspondiente (del 1 al 10)
# Usamos un "list comprehension" para que el número quede justo en el medio de la barra
plt.xticks([i + 0.5 for i in range(1, 11)], labels=range(1, 11))

plt.grid(axis='y', alpha=0.3, linestyle='--')

plt.show()
```

🎯 Ejercicio 2: El Histograma (Histplot)
La pregunta a responder: ¿Cómo se distribuyen los niveles de esfuerzo de las alumnas? ¿La mayoría hace un esfuerzo bajo, medio o alto?

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Cargar los datos y calcular las medias
df = pd.read_csv('/content/desempeno_fisico.csv')
esfuerzo_promedio = df.groupby('estudiante')['esfuerzo'].mean()
esfuerzo_promedio_alumnas = esfuerzo_promedio.mean()

print("Promedio por estudiante:\n", esfuerzo_promedio)
print("\nPromedio general de las alumnas:", esfuerzo_promedio_alumnas)

# 2. Definir los umbrales (puedes ajustar estos porcentajes)
# "Muy por debajo" será, por ejemplo, menos del 80% de la media general
umbral_muy_bajo = esfuerzo_promedio_alumnas * 0.8

# 3. Crear la lista de colores dinámicamente
color_barras = []
for valor in esfuerzo_promedio.values:
    if valor >= esfuerzo_promedio_alumnas:
        color_barras.append('green')       # Verde: Por encima o igual a la media
    elif valor < esfuerzo_promedio_alumnas and valor >= umbral_muy_bajo:
        color_barras.append('orange')      # Naranja: Por debajo de la media
    else:
        color_barras.append('red')         # Rojo: Muy por debajo de la media

# 4. Dibujar el gráfico
plt.figure(figsize=(10, 6)) # Añadido para que se vea mejor si hay muchos estudiantes
plt.bar(esfuerzo_promedio.index, esfuerzo_promedio.values, color=color_barras)

# Añadir una línea discontinua que marque la media general para que se entienda mejor el gráfico
plt.axhline(y=esfuerzo_promedio_alumnas, color='blue', linestyle='--', label=f'Media general ({esfuerzo_promedio_alumnas:.2f})')

plt.title('Esfuerzo promedio por estudiante')
plt.xlabel('Estudiante')
plt.ylabel('Esfuerzo promedio')
plt.xticks(rotation=45) # Rota los nombres si son largos para que no se solapen
plt.legend()
plt.tight_layout() # Ajusta los márgenes automáticamente

plt.show()
```

🗺️ Ejercicio 3: El Gráfico de Dispersión (Scatterplot)
La pregunta a responder: ¿Las alumnas que tardan más tiempo en una actividad son también las que hacen más esfuerzo? ¿Cambia esto según el tipo de actividad?

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # Asegúrate de tener seaborn importado

# 1. Cargar los datos
df = pd.read_csv('/content/desempeno_fisico.csv')

# 2. Configurar el tamaño del gráfico
plt.figure(figsize=(10, 6))

# 3. Crear el gráfico de dispersión
# Usamos 'hue' para diferenciar las actividades por colores y 'style' por si quieres formas distintas
sns.scatterplot(
    data=df,
    x='tiempo',
    y='esfuerzo',
    hue='actividad',
    style='actividad',
    s=100, # Tamaño de los puntos
    alpha=0.8 # Transparencia para ver si hay puntos solapados
)

# 4. Personalizar el gráfico
plt.title('Relación entre Tiempo Invertido y Esfuerzo por Actividad', fontsize=14, fontweight='bold')
plt.xlabel('Tiempo en la actividad (minutos/horas)', fontsize=12)
plt.ylabel('Esfuerzo realizado', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5) # Añade una cuadrícula sutil de fondo
plt.legend(title='Tipo de Actividad', bbox_to_anchor=(1.05, 1), loc='upper left') # Mueve la leyenda fuera para que no estorbe

# 5. Mostrar el gráfico
plt.tight_layout()
plt.show()
```

📦 Ejercicio 4: El Gráfico de Caja (Boxplot)
La pregunta a responder: ¿En qué actividad física hay mayor variación o dispersión en los tiempos de las alumnas?

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Cargar los datos
df = pd.read_csv('/content/desempeno_fisico.csv')

# 2. Configurar el tamaño del gráfico
plt.figure(figsize=(10, 6))

# 3. Crear el gráfico de caja (Boxplot)
# X será la actividad (para comparar los grupos) e Y será el tiempo
sns.boxplot(
    data=df,
    x='actividad',
    y='tiempo',
    palette='Set2' # Aplica una paleta de colores bonita y clara
)

# 4. Personalizar el gráfico
plt.title('Dispersión del Tiempo por Tipo de Actividad Física', fontsize=14, fontweight='bold')
plt.xlabel('Tipo de Actividad', fontsize=12)
plt.ylabel('Tiempo invertido', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.3) # Una cuadrícula de fondo muy sutil

# 5. Mostrar el gráfico
plt.tight_layout()
plt.show()
```

📈 Ejercicio 5: El Gráfico de Líneas (Lineplot)
La pregunta a responder: ¿Cómo cambia el esfuerzo promedio a medida que aumenta la edad de las alumnas?

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Cargar los datos
df = pd.read_csv('/content/desempeno_fisico.csv')

# 2. Configurar el tamaño del gráfico
plt.figure(figsize=(10, 5))

# 3. Crear el gráfico de líneas
# X es la variable que avanza (Edad) e Y es lo que queremos medir (Esfuerzo)
sns.lineplot(
    data=df,
    x='edad',
    y='esfuerzo',
    marker='o',          # Añade un punto en cada edad para ver los datos exactos
    linewidth=2.5,       # Hace la línea un poco más gruesa
    color='b'            # Color azul para la línea
)

# 4. Personalizar el gráfico
plt.title('Evolución del Esfuerzo Promedio según la Edad', fontsize=14, fontweight='bold')
plt.xlabel('Edad de las alumnas', fontsize=12)
plt.ylabel('Esfuerzo promedio', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5) # Cuadrícula para seguir mejor la línea

# 5. Mostrar el gráfico
plt.tight_layout()
plt.show()
```
