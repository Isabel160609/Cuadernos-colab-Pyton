```python
import matplotlib.pyplot as plt



# 2. Definimos una lista con un color diferente para cada una de las 5 actividades
# Puedes usar nombres de colores en inglés o códigos HEX (como #FF5733)
colores_personalizados = ['skyblue', 'salmon', 'lightgreen', 'gold', 'orchid',"pink", "orange","brown", "red", "purple"]

# 3. Creamos el gráfico de quesito
plt.figure(figsize=(6, 6)) # Hacemos la ventana cuadrada para que el queso sea un círculo perfecto
felicidad_region_media = df.groupby('Region')['Happiness Rank'].mean()
print(felicidad_region_media)
porcentaje_felicidad_region = felicidad_region_media / felicidad_region_media.sum() * 100
print(porcentaje_felicidad_region)
plt.pie(
    felicidad_region_media,                      # Los valores numéricos (el tamaño de cada porción)
    labels=felicidad_region_media.index,         # Los nombres de cada porción
    colors=colores_personalizados, # ¡Aquí asignamos los colores diferentes!
    autopct='%1.1f%%',          # Esto calcula y pone el porcentaje automáticamente
    startangle=140,             # Rota el gráfico para una mejor perspectiva visual
    shadow = True,
    wedgeprops={'edgecolor': 'white', 'linewidth': 2} # Añade una línea blanca elegante entre los quesitos
)

# 4. Título del gráfico
plt.title("Felicidad segun Region", fontsize=14, fontweight='bold')

# 5. Mostramos el resultado
plt.show()
```
