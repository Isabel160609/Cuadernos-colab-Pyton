El Inventario de Frutas
Enunciado: Crea un diccionario llamado frutas donde las claves sean nombres de frutas y los valores sean su cantidad en stock.

1-Añade: "manzanas" con valor 10 y "bananas" con valor 5.

2-Actualiza el stock de "bananas" a 12.

3-Añade una nueva fruta "naranjas" con valor 20.

4-Imprime el stock total de "manzanas".

```python
frutas = {"manzanas": 10, "bananas": 5}
frutas["bananas"]=12
frutas["naranjas"]=20
print(f"Stock de manzanas: {frutas['manzanas']}") #accedemos al valor de la clave "manzanas"
print(f"Inventario completo: {frutas}")
```
